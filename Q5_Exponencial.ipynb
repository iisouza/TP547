Importando as bibliotecas necessárias:

```python
import numpy as np
import matplotlib.pyplot as plt
```

O número de dias que os viajantes compram com antecedência suas passagens aéreas pode ser modelado por uma Distribuiçãoo Exponencial com o tempo médio igual a 28 dias. Encontre a probabilidade de um viajante comprar uma passagem com menos de 4 dias de antecedência.

Inicialização das Variáveis segundo a Distribuição de Poisson:

```python
N = 100000          # Número de amostras.
lambda1 = 1/28      # Média segundo Poisson.
x = np.random.uniform(0, 1, N)
```

```python
value, count = [4, 0]
Xexp = -np.log(x)/lambda1

for expvalue in Xexp:
    if expvalue <= value:
        count += 1

print('Probabilidade de menos de 4 dias de antecedência = {:.4}'.format(count/N))
```

Traçar a Função Densidade de Probabilidade da Variável Analisada.

```python
X=np.arange(0, 200, 0.1)
fx = lambda1*np.exp(-lambda1*X)
plt.plot(X, fx)
plt.hist(Xexp, bins=100, density=True)

plt.title('Função Densidade de Probabilidade')
plt.show()
```
