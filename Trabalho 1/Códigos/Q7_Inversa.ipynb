Importando as bibliotecas necessárias:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Utilizando o método da inversa, gerar amostras para a distribuição

$$f(x)=\frac{e^x}{e^2-1},\ 0\leq x\leq 2$$

```python
n = 100000
x_values = np.random.uniform(0, 1, n)
x_function = np.log((np.exp(2) - 1)*x_values + 1)
```

Histograma das amostras geradas:

```python
X=np.arange(0, 2, 0.01)
fx = np.exp(X)/(np.exp(2)-1)
plt.plot(X, fx)

plt.hist(x_function, bins=50, density=True)
plt.title('Amostras da Distribuição f(x)')
plt.show()
```
