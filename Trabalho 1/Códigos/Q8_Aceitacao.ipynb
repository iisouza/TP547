Importando as bibliotecas necessárias:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Utilizando o método da aceitação/rejeição, gerar amostras para a distribuição

$$f(x) = 1.5x^2, −1 < x < 1$$

```python
n = 10000
xfunction = np.array([])

for i in range(n):
    x1 = np.random.uniform(-1, 1)   # Amostras no intervalo [-1, 1]
    x2 = np.random.uniform(0, 1.5)  # Amostras no intervalo [0, 1.5]

    while (x2 >= 1.5 * x1 ** 2):    # Critério de aceitação/rejeição
        x1 = np.random.uniform(-1, 1)
        x2 = np.random.uniform(0, 1.5)
    xfunction = np.append(xfunction, x1)

X=np.arange(-1, 1, 0.01)
fx = 1.5 * X ** 2
plt.plot(X, fx)
plt.hist(xfunction, bins=100, density=True)
plt.show()
```
